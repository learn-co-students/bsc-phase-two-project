**Created a function to take in the column 'zipcode' and return the name of the school district it is coded for. \
This is not imperative to our model, but it was useful for our vision**

```python
def sch_district(zipcode):
    zipcode = int()
    if zipcode == 98040:
        return 'Mercer Island'
    if zipcode == 98004 or 98005 or 98006 or 98007 or 98008 or 98039:
        return 'Bellevue'
    if zipcode == 98052:
        return 'Lake Washington'
    if zipcode == 98027 or 98029:
        return 'Issaquah'
    if zipcode == 98038 or 98042:
        return 'Tahoma'
    if zipcode == 98115:
        return 'Shoreline'
    if zipcode == 98070:
        return 'Vashon'
    if zipcode == 98065 or 98068 or 98024:
        return 'Snoqualmie'
    if zipcode == 98124 or 98117 or 98148:
        return 'Seattle'
    if zipcode == 98019:
        return 'Riverview'
    if zipcode == 98002 or 98092 or 98001:
        return 'Auburn'
    if zipcode == 98022:
        return 'Enumclaw'
    if zipcode == 98288:
        return 'Skykomish'
    if zipcode == 98028 or 98032 or 98010:
        return 'Kent'
    if zipcode == 98055 or 98056 or 98057 or 98059:
        return 'Renton'
    if zipcode == 98003 or 98023:
        return 'Federal Way'
    if zipcode == 98116:
        return 'Highline'
    if zipcode == 98118:
        return 'Tukwila'
    if zipcode == 98092:
        return 'Muckleshoot'
    else:
        return 'Zipcode is not in range of King County School Districts'
```

**Created a function to take in the column 'zipcode' and return the name of the school district it is coded for. \
This is not imperative to our model, but it was useful for our vision**

```python
def sch_dist_grade(zipcode):
    zipcode_c = int(zipcode)
    if zipcode_c == 98040 or 98004 or 98005 or  98006 or  98007 or  98008or  98052 or  98027 or  98029 or  98039:
        return 'A+'
    if zipcode_c == 98038 or  98042 or  98115 or  98070:
        return 'A'
    if zipcode_c == 98065 or  98068 or 98101 or 98102 or 98103 or 98124 or 98125 or 98024 or 98148:
        return 'A-'
    if zipcode_c == 98019:
        return 'B+'
    if zipcode_c == 98002 and 98092 or 98022 or 98288 or 98028 or 98032 or 98001 or 98010:
        return 'B'
    if zipcode_c == 98052 or 98055 or 98056 or 98057 or 98059:
        return 'B-'
    if zipcode_c == 98003 or 98023 or 98116 or 98118:
        return 'C+'
    if zipcode_c == 98092:
        return 'D+'
    else:
        return 0
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression        
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy import asarray
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_squared_error as mse
from scipy.stats import zscore
from scipy.stats import stats
```

```python
reg = LinearRegression()
y = df['price']
X = df.drop(['id', 'price'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
encoder = OneHotEncoder(sparse=False)
encoder.fit(X_train[['view', 'grade', 'zipcode']])
transformed_train = encoder.transform(X_train[['view', 'grade', 'zipcode']])
transformed_train = pd.DataFrame(transformed_train, columns = encoder.get_feature_names(), index=X_train.index)
X_train_encoded = pd.concat([X_train.drop(['view', 'grade', 'zipcode'], axis = 1), transformed_train], axis = 1)
test_condition = encoder.transform(X_test[['view', 'grade', 'zipcode']])
test_condition = pd.DataFrame(test_condition, columns=encoder.get_feature_names(), index=X_test.index)
test_condition = pd.concat([X_test.drop(['view', 'grade', 'zipcode'], axis = 1), test_condition], axis = 1)
```

```python
reg.fit(X_train_encoded, y_train)
reg.score(X_train_encoded, y_train)
```

```python
reg.score(test_condition, y_test)
```

```python
train_preds = reg.predict(X_train_encoded)
test_preds = reg.predict(test_condition)
```

```python
np.sqrt(mse(y_train, train_preds))
```

```python
mse(y_test, test_preds, squared=False)
```
